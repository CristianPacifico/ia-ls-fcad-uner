Uno de los datasets más famosos del mundo del Machine Learning: el **Titanic**. A partir de datos de los pasajeros, queremos predecir si una persona **sobrevivió (1)** o **no (0)**.

Compararemos dos modelos:

| Modelo | Idea principal |
|---|---|
| 🌳 **Decision Tree** | Divide datos con preguntas binarias (ej: ¿es masculino? ¿paga más de X?) |
| 📈 **Logistic Regression** | Calcula la probabilidad de sobrevivir usando una función sigmoide |

## 📦 1. Importación de librerías

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

## 📂 2. Carga y preparación del dataset

Cargamos el dataset del Titanic y creamos una feature booleana `male` a partir de la columna `Sex`.

Los modelos de ML no entienden directamente texto como `'male'` o `'female'`, por eso lo convertimos a `True/False` (que Python trata como `1/0`).

```python
df = pd.read_csv('https://raw.githubusercontent.com/CristianPacifico/ia-ls-fcad-uner/main/data/ml/titanic.csv')

# Convertimos 'Sex' en una variable binaria
df['male'] = df['Sex'] == 'male'

print("🔍 Primeras filas del dataset:")
print(df.head())
print(f"\n📐 Estructura del dataset: {df.shape[0]} filas x {df.shape[1]} columnas")
```

## ✂️ 3. Definición de features y separación de datos

**Features seleccionadas:**
- `Pclass`: Clase del pasajero (1ª, 2ª o 3ª clase)
- `male`: ¿Es hombre? (True/False)
- `Age`: Edad
- `Siblings/Spouses`: Hermanos/cónyuge a bordo
- `Parents/Children`: Padres/hijos a bordo
- `Fare`: Precio del pasaje

**Target:** `Survived` → 1 (sobrevivió) / 0 (no sobrevivió)

```python
feature_cols = ['Pclass', 'male', 'Age', 'Siblings/Spouses', 'Parents/Children', 'Fare']
X = df[feature_cols].values   # Variables independientes
y = df['Survived'].values     # Variable objetivo

# División: 70% entrenamiento, 30% testeo
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=105
)

print(f"🏋️  Entrenamiento: {X_train.shape[0]} muestras")
print(f"🧪  Testeo:        {X_test.shape[0]} muestras")
```

## 🏗️ 4. Creación y entrenamiento de los modelos

```python
# Creación de modelos
modelDT = DecisionTreeClassifier()
modelLR = LogisticRegression()

# Entrenamiento
modelDT.fit(X_train, y_train)
modelLR.fit(X_train, y_train)

print("✅ Modelos entrenados exitosamente")
```

## 🔮 5. Predicciones sobre datos de testeo

Veamos las primeras predicciones y comparémoslas con los valores reales.

```python
y_predDT = modelDT.predict(X_test)
y_predLR = modelLR.predict(X_test)

print("🌳 Primeras 20 predicciones DT:", y_predDT[:20])
print("📈 Primeras 20 predicciones LR:", y_predLR[:20])
print("🎯  Primeros 20 valores reales:", y_test[:20])
```

## 📊 6. Evaluación de accuracy

Calculamos la **exactitud** de cada modelo de tres maneras equivalentes:

1. Contando manualmente los aciertos
2. Calculando el promedio de aciertos
3. Usando `.score()` de scikit-learn

```python
print("\n" + "="*50)
print("  🌳 DECISION TREE")
print("="*50)
print(f"   Total de aciertos: {(y_test == y_predDT).sum()} de {len(y_predDT)}")
print(f"  Promedio de aciertos: {(y_test == y_predDT).sum() / y_test.shape[0]:.4f}")
print(f"  Score / Accuracy: {modelDT.score(X_test, y_test):.4f}")

print("\n" + "="*50)
print("  📈 LOGISTIC REGRESSION")
print("="*50)
print(f"   Total de aciertos: {(y_test == y_predLR).sum()} de {len(y_predLR)}")
print(f"  Promedio de aciertos: {(y_test == y_predLR).sum() / y_test.shape[0]:.4f}")
print(f"  Score / Accuracy: {modelLR.score(X_test, y_test):.4f}")
```

## 🎨 7. Visualización: Accuracy comparativo

```python
import matplotlib.pyplot as plt

modelos = ['Decision Tree', 'Logistic Regression']
accuracies = [modelDT.score(X_test, y_test), modelLR.score(X_test, y_test)]

colors = ['steelblue', 'darkorange']
fig, ax = plt.subplots(figsize=(7, 4))
bars = ax.bar(modelos, accuracies, color=colors, width=0.4)
ax.set_ylim(0, 1.1)
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy: Decision Tree vs Logistic Regression\nDataset Titanic')
ax.bar_label(bars, fmt='%.4f', padding=5, fontsize=12)
plt.tight_layout()
plt.show()
```

## 🏁 8. Conclusiones

- 🚢 En el dataset Titanic, la Regresión Logística suele rendir similar o mejor que un Árbol de Decisión sin restricciones, ya que el DT puede sobreajustarse.
- 🔑 La variable `male` (género) y `Pclass` (clase) son predictores muy influyentes en la supervivencia.
- 📉 Para el DT, ajustar `max_depth` puede mejorar la generalización.
