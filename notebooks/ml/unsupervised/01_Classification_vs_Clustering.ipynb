## 🧠 Marco Teórico

### 📋 ¿Qué es la Clasificación?

La **clasificación** es una tarea de *aprendizaje supervisado*: utiliza datos **etiquetados** (pares entrada-salida conocidos) para aprender una función que predice la clase discreta de nuevas instancias.

$$f: X \rightarrow y_{\text{discreta}}$$

**Casos de uso:**
- 📧 **Filtrado de spam** — ¿Es un correo spam o no?
- 👤 **Reconocimiento facial** — ¿A quién pertenece esta cara?
- 💳 **Aprobación de préstamos** — ¿El solicitante es elegible?
- 📉 **Predicción de churn** — ¿Este cliente va a cancelar su suscripción?

---

#### Algoritmos de Clasificación

##### Regresión Logística

A pesar de su nombre, la regresión logística es un algoritmo de **clasificación**. Estima la probabilidad de pertenencia a una clase usando la función logística (sigmoide). Una frontera de decisión separa las clases.

![Función Logística — Scikit-Learn](https://media.datacamp.com/legacy/v1694516402/image5_bec2398d66.png)

*Fuente: [Logistic Function — Scikit-Learn](https://scikit-learn.org/stable/auto_examples/linear_model/plot_logistic.html). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

---

##### K-Nearest Neighbors (KNN)

KNN es un algoritmo **no paramétrico** y de **aprendizaje perezoso** (*lazy learner*): no construye un modelo explícito. Clasifica nuevas instancias por mayoría de votos entre sus $k$ vecinos más cercanos en el espacio de features.

![KNN — Scikit-Learn](https://media.datacamp.com/legacy/v1694516490/image9_ef51e86a00.png)

*Fuente: [K-Nearest Neighbor Classification — DataCamp](https://www.datacamp.com/tutorial/k-nearest-neighbor-classification-scikit-learn). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

---

##### Árboles de Decisión

Aprendizaje de reglas en forma de árbol: **no paramétrico**, intuitivo y fácilmente interpretable. El camino de la raíz a una hoja define una regla de decisión sobre las features.

![Árbol de Decisión](https://media.datacamp.com/legacy/v1694516555/image1_b349ad4721.png)

*Fuente: [Decision Tree Classification in Python — DataCamp](http://datacamp.com/tutorial/decision-tree-classification-python). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

---

##### Random Forest

Modelo **ensemble** de múltiples árboles de decisión. Combina **bagging** (muestreo con reemplazo de instancias) con el método del subespacio aleatorio (*feature bagging*) para reducir la correlación entre los predictores individuales y mejorar la generalización.

![Bagging — Random Forest](https://media.datacamp.com/legacy/v1694516613/image2_e56597ba90.png)

*Fuente: [Bagging: Machine Learning through Visuals](https://medium.com/machine-learning-through-visuals/machine-learning-through-visuals-part-1-what-is-bagging-ensemble-learning-432059568cc8). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

---

##### Naïve Bayes

Clasificador **probabilístico** basado en el teorema de Bayes. Asume que todas las features son **independientes** entre sí (suposición *naive*), lo que simplifica el cálculo manteniéndolo competitivo en muchos problemas reales:

$$P(C|X) = \frac{P(X|C) \cdot P(C)}{P(X)}$$

![Naïve Bayes](https://media.datacamp.com/legacy/v1694516692/image4_b3950ce88b.png)

*Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

### 🔵 ¿Qué es el Clustering?

El **clustering** es una técnica de *aprendizaje no supervisado*: trabaja con datos **sin etiquetas** y descubre la estructura subyacente agrupando instancias similares. No hay un "maestro" que indique las respuestas correctas.

**Casos de uso:**
- 🏪 **Segmentación de mercado** — Agrupar clientes con hábitos similares
- 📷 **Segmentación de imágenes** — Particionar imágenes en regiones significativas
- 🌐 **Análisis de redes sociales** — Detectar comunidades en una red
- 🛒 **Motores de recomendación** — Descubrir perfiles de comportamiento

---

#### Algoritmos de Clustering

##### K-Means

El más popular. Es **centroide-based** e **iterativo**: asigna cada punto al centroide más cercano y actualiza los centroides hasta convergencia. Requiere especificar $k$ (número de clusters) de antemano.

![K-Means Clustering](https://media.datacamp.com/legacy/v1694516799/image3_bdac8621d1.png)

*Fuente: [LearnByMarketing](https://www.learnbymarketing.com/wp-content/uploads/2015/01/method-k-means-steps-example.png). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

---

##### Clustering Jerárquico — Aglomerativo (bottom-up)

Comienza con cada punto como su propio cluster y los fusiona en pares según distancia, construyendo una jerarquía completa visualizable como **dendrograma**.

![Clustering Jerárquico Aglomerativo](https://media.datacamp.com/legacy/v1694516908/image7_0bfd55885f.png)

*Fuente: [Introduction to Hierarchical Clustering — DataCamp](https://www.datacamp.com/tutorial/introduction-hierarchical-clustering-python). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

##### Clustering Jerárquico — Divisivo (top-down)

Enfoque opuesto: parte de un único cluster que contiene todos los puntos y realiza divisiones recursivas.

![Clustering Jerárquico Divisivo](https://media.datacamp.com/legacy/v1694516957/image8_9a9eaa4c3e.png)

*Fuente: [Introduction to Hierarchical Clustering — DataCamp](https://www.datacamp.com/tutorial/introduction-hierarchical-clustering-python). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

---

##### DBSCAN

*Density-Based Spatial Clustering of Applications with Noise*. Agrupa puntos en regiones **densas** separadas por zonas de baja densidad. Sus ventajas clave:
- No requiere especificar $k$ de antemano
- Detecta clusters de **forma arbitraria**
- Es robusto ante **outliers** (los marca como ruido: etiqueta $-1$)

![DBSCAN](https://media.datacamp.com/legacy/v1694517021/image6_560b21e635.png)

*Fuente: [DBSCAN macroscopic investigation — DataCamp](https://www.datacamp.com/tutorial/dbscan-macroscopic-investigation-python). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

---

##### OPTICS

*Ordering Points To Identify the Clustering Structure*. Desarrollado por el mismo grupo que DBSCAN, supera su limitación principal: funciona correctamente con datos de **densidad variable**.

![OPTICS — Scikit-Learn](https://media.datacamp.com/legacy/v1694517128/image11_a076908ce1.png)

*Fuente: [Demo of OPTICS — Scikit-Learn](https://scikit-learn.org/stable/auto_examples/cluster/plot_optics.html). Imagen incluida en [Pykes, K. (2023). Classification vs Clustering — DataCamp](https://www.datacamp.com/blog/classification-vs-clustering-in-machine-learning).*

---

### ⚡ Diferencias Clave

| | Clasificación | Clustering |
|---|---|---|
| **Tipo de aprendizaje** | Supervisado | No supervisado |
| **Requiere etiquetas** | ✅ Sí | ❌ No |
| **Propósito** | Predecir clase de nuevas instancias | Descubrir grupos naturales en los datos |
| **Evaluación** | Accuracy, F1, ROC-AUC | Silhouette, ARI, Inercia |
| **Split train/test** | Necesario | No obligatorio |
| **Algoritmos** | LR, KNN, DT, RF, Naive Bayes | K-Means, Hierarchical, DBSCAN, OPTICS |
| **Casos de uso** | Spam, churn, diagnóstico | Segmentación, recomendación, anomalías |

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

from sklearn import datasets, metrics
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, cross_val_score

# Algoritmos de Clasificación
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

# Algoritmos de Clustering
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.datasets import make_moons

SEED = 42
np.random.seed(SEED)

print('✅ Librerías importadas correctamente!')
```

## 🏷️ Paso 2: Clasificación — Aprendiendo con Datos Etiquetados

Usamos el clásico dataset **Iris** (150 muestras, 4 features, 3 clases: Setosa, Versicolor, Virginica) para comparar 5 algoritmos de clasificación.

En clasificación:
1. Los datos tienen **etiquetas** conocidas (sabemos a qué especie pertenece cada flor)
2. Se divide en entrenamiento y test
3. El modelo **aprende** de los pares (features → clase)
4. Se evalúa la capacidad de **predicción** en datos nuevos


```python
# Cargar dataset Iris
iris = datasets.load_iris()
X, y = iris.data, iris.target

print('🌸 Dataset Iris:')
print(f'  Muestras: {X.shape[0]}')
print(f'  Features: {X.shape[1]}  →  {iris.feature_names}')
print(f'  Clases:   {list(iris.target_names)}')
print(f'  Distribución: {dict(zip(iris.target_names, np.bincount(y)))}')
print()
print('  ✅ Las etiquetas (y) son CONOCIDAS. Esta es la clave del aprendizaje supervisado.')
print('  Sin estas etiquetas, solo tendríamos clustering.')
```

```python
# Dividir en entrenamiento (70%) y test (30%) con estratificación
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=SEED, stratify=y)

# Estandarizar (ajustar solo sobre train → sin data leakage)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)

# Comparar 5 algoritmos con Cross-Validation (CV=5) sobre el conjunto de entrenamiento
classifiers = {
    'Logistic Regression': LogisticRegression(random_state=SEED, max_iter=1000),
    'KNN (k=5)':           KNeighborsClassifier(n_neighbors=5),
    'Decision Tree':       DecisionTreeClassifier(random_state=SEED, max_depth=4),
    'Random Forest':       RandomForestClassifier(n_estimators=100, random_state=SEED),
    'Naïve Bayes':         GaussianNB(),
}

resultados_clf = {}
print('📊 Accuracy CV=5 en entrenamiento:')
print(f"  {'Algoritmo':<22} {'Accuracy':>10}  {'±Std':>7}")
print('  ' + '-' * 44)
for nombre, clf in classifiers.items():
    scores = cross_val_score(clf, X_train_s, y_train, cv=5, scoring='accuracy')
    resultados_clf[nombre] = (scores.mean(), scores.std())
    print(f"  {nombre:<22} {scores.mean()*100:>9.1f}%  {scores.std()*100:>6.1f}%")

print()
print('📊 Accuracy final en TEST (datos no vistos):')
print(f"  {'Algoritmo':<22} {'Test Acc':>10}")
print('  ' + '-' * 35)
for nombre, clf in classifiers.items():
    clf.fit(X_train_s, y_train)
    print(f"  {nombre:<22} {clf.score(X_test_s, y_test)*100:>9.1f}%")
```

```python
# Gráfico de comparación de algoritmos de clasificación
fig, ax = plt.subplots(figsize=(10, 5))

nombres  = list(resultados_clf.keys())
medias   = [v[0] * 100 for v in resultados_clf.values()]
stds     = [v[1] * 100 for v in resultados_clf.values()]
colores  = ['#4C72B0', '#DD8452', '#55A868', '#C44E52', '#8172B2']

bars = ax.barh(nombres, medias, xerr=stds, color=colores,
               edgecolor='black', alpha=0.85,
               error_kw={'elinewidth': 2, 'capsize': 5})

for bar, media in zip(bars, medias):
    ax.text(media + 0.5, bar.get_y() + bar.get_height() / 2,
            f'{media:.1f}%', va='center', fontsize=10)

ax.set_xlabel('Accuracy CV (%)', fontsize=12)
ax.set_title('Clasificación sobre Iris — Comparación de Algoritmos (CV=5)', fontsize=13)
ax.set_xlim(80, 107)
ax.axvline(90, color='gray', linestyle='--', alpha=0.4, label='90%')
ax.grid(True, axis='x', alpha=0.3)
ax.legend(fontsize=9)
plt.tight_layout()
plt.show()

print('💡 Clasificación = se conocen las respuestas (etiquetas) durante el entrenamiento.')
print('   El modelo aprende de esos ejemplos etiquetados para predecir nuevos casos.')
```

## 🔵 Paso 3: Clustering K-Means — Descubriendo Grupos Sin Etiquetas

Ahora usamos los **mismos datos de Iris**, pero **ignoramos las etiquetas**. K-Means no sabe cuántas clases hay ni cuáles son; solo recibe las features y busca $k$ grupos naturales basándose en similitud.

**Para evaluar** la calidad del clustering (ya que tenemos las etiquetas verdaderas disponibles), usaremos el **Adjusted Rand Index (ARI)**:
- ARI = 1.0 → los clusters coinciden perfectamente con las clases reales
- ARI ≈ 0.0 → resultado equivalente a asignación aleatoria

> 🔑 Importante: el ARI solo se calcula *a posteriori* para validación. El algoritmo de clustering **no lo usó** durante el proceso.


```python
# Escalado de Iris (sin usar las etiquetas y)
X_esc = StandardScaler().fit_transform(X)

# Reducción a 2D con PCA para visualización
pca = PCA(n_components=2, random_state=SEED)
X_pca = pca.fit_transform(X_esc)

# K-Means con k=3
kmeans = KMeans(n_clusters=3, random_state=SEED, n_init=10)
labels_km = kmeans.fit_predict(X_esc)

# Proyectar los centroides al espacio PCA
cent_pca = pca.transform(kmeans.cluster_centers_)

# Evaluar cuán bien K-Means recuperó la estructura real
ari_km        = metrics.adjusted_rand_score(y, labels_km)
silhouette_km = metrics.silhouette_score(X_esc, labels_km)

print('🔵 K-Means sobre Iris (sin etiquetas, k=3):')
print(f'  Adjusted Rand Index (ARI):  {ari_km:.3f}   [0=aleatorio, 1=perfecto]')
print(f'  Silhouette Score:           {silhouette_km:.3f}  [-1=mal, 0=solapado, 1=ideal]')
print()
print('  El ARI nos dice cuánto se parecen los clusters encontrados a las clases reales.')
print('  El Silhouette mide cuán separados están los clusters entre sí.')

# Colores y marcadores para reutilizar en celdas siguientes
colores_clases   = ['royalblue', 'tomato', 'mediumseagreen']
colores_clusters = ['purple', 'darkorange', 'teal']
formas           = ['o', 's', '^']
```

```python
# Comparación visual: clases reales (clasificación) vs clusters K-Means (clustering)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Iris — Proyección PCA 2D: Clases Reales vs Clusters K-Means', fontsize=13)

# ── Panel izquierdo: clases reales ──
ax = axes[0]
for i, nombre_clase in enumerate(iris.target_names):
    mask = y == i
    ax.scatter(X_pca[mask, 0], X_pca[mask, 1],
               c=colores_clases[i], marker=formas[i],
               label=nombre_clase, s=70,
               edgecolors='black', linewidths=0.5, alpha=0.9)
ax.set_title('CLASIFICACIÓN — Clases REALES (etiquetadas)', fontsize=11)
ax.set_xlabel('PCA Componente 1')
ax.set_ylabel('PCA Componente 2')
ax.legend(title='Clase', fontsize=9)
ax.grid(True, alpha=0.3)
ax.text(0.05, 0.95, '✅ Supervisado\n(usa etiquetas)', transform=ax.transAxes,
        fontsize=9, va='top',
        bbox={'boxstyle': 'round', 'facecolor': 'lightblue', 'alpha': 0.6})

# ── Panel derecho: clusters K-Means ──
ax = axes[1]
for cluster_id in range(3):
    mask = labels_km == cluster_id
    ax.scatter(X_pca[mask, 0], X_pca[mask, 1],
               c=colores_clusters[cluster_id], marker=formas[cluster_id],
               label=f'Cluster {cluster_id}', s=70,
               edgecolors='black', linewidths=0.5, alpha=0.9)
ax.scatter(cent_pca[:, 0], cent_pca[:, 1],
           c='black', marker='x', s=200, linewidths=3,
           label='Centroides', zorder=5)
ax.set_title(f'CLUSTERING K-Means — Grupos DESCUBIERTOS (ARI={ari_km:.2f})', fontsize=11)
ax.set_xlabel('PCA Componente 1')
ax.set_ylabel('PCA Componente 2')
ax.legend(title='Cluster', fontsize=9)
ax.grid(True, alpha=0.3)
ax.text(0.05, 0.95, '🔵 No Supervisado\n(sin etiquetas)', transform=ax.transAxes,
        fontsize=9, va='top',
        bbox={'boxstyle': 'round', 'facecolor': 'lightyellow', 'alpha': 0.6})

plt.tight_layout()
plt.show()

print('💡 Setosa (cluster inferior) es fácilmente separable en ambos casos.')
print('   Versicolor y Virginica se solapan → K-Means comete algunos errores.')
print(f'   Aún así, con ARI={ari_km:.2f} recuperó bien la estructura general de los datos.')
```

## 🌳 Paso 4: Clustering Jerárquico — El Dendrograma

El clustering jerárquico **no requiere especificar el número de clusters** de antemano: construye una jerarquía completa que puede "cortarse" a cualquier nivel para obtener distintos números de clusters.

La representación visual característica es el **dendrograma**: cada fusión (aglomerativo) aparece como una rama. La altura de la rama indica la distancia a la que se fusionaron los clusters.


```python
# Dendrograma sobre una muestra de 30 instancias (para legibilidad del gráfico)
indices_muestra = np.random.choice(len(X), 30, replace=False)
X_muestra = X_esc[indices_muestra]
y_muestra = y[indices_muestra]

# Calcular linkage con método Ward (minimiza la varianza intra-cluster)
Z = linkage(X_muestra, method='ward')

fig, ax = plt.subplots(figsize=(15, 5))
ax.set_title('Dendrograma — Clustering Jerárquico (Ward) sobre Iris [muestra de 30 puntos]',
             fontsize=13)

# Etiquetas con la clase real (solo para verificación post-hoc)
etiquetas = [iris.target_names[yi] for yi in y_muestra]

dendrogram(Z, labels=etiquetas, ax=ax,
           color_threshold=5, leaf_font_size=9, leaf_rotation=45)
ax.axhline(y=5, color='red', linestyle='--', alpha=0.7, label='Umbral de corte → 3 clusters')
ax.set_xlabel('Instancia')
ax.set_ylabel('Distancia de Ward')
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()

print('💡 Cortando el dendrograma en el umbral rojo, obtenemos ~3 grupos.')
print('   Las etiquetas de clase se muestran solo para verificar la calidad del resultado.')
print('   El algoritmo NO las usó para construir la jerarquía.')
```

```python
# AgglomerativeClustering sobre TODO el dataset Iris (n_clusters=3)
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels_agg = agg.fit_predict(X_esc)

ari_agg  = metrics.adjusted_rand_score(y, labels_agg)
silh_agg = metrics.silhouette_score(X_esc, labels_agg)

print('🌳 Comparación K-Means vs Clustering Jerárquico (Iris, 3 clusters):')
print(f"  {'Algoritmo':<25} {'ARI':>8} {'Silhouette':>12}")
print('  ' + '-' * 48)
print(f"  {'K-Means':<25} {ari_km:>8.3f} {silhouette_km:>12.3f}")
print(f"  {'Jerárquico (Ward)':<25} {ari_agg:>8.3f} {silh_agg:>12.3f}")

# Visualización del resultado aglomerativo
fig, ax = plt.subplots(figsize=(7, 5))
for cluster_id in range(3):
    mask = labels_agg == cluster_id
    ax.scatter(X_pca[mask, 0], X_pca[mask, 1],
               c=colores_clusters[cluster_id], marker=formas[cluster_id],
               label=f'Cluster {cluster_id}', s=70,
               edgecolors='black', linewidths=0.5, alpha=0.9)
ax.set_title(f'Clustering Jerárquico Aglomerativo — Iris (ARI={ari_agg:.2f})', fontsize=12)
ax.set_xlabel('PCA Componente 1')
ax.set_ylabel('PCA Componente 2')
ax.legend(title='Cluster')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 🌌 Paso 5: DBSCAN — Clusters de Forma Arbitraria

K-Means asume que los clusters son **esféricos** y de **tamaño similar**. DBSCAN supera esta limitación: puede encontrar clusters de **cualquier forma** y detectar **outliers** automáticamente (sin necesitar que el usuario los defina).

Demostramos esto con el dataset `make_moons` (dos lunas entrelazadas), donde K-Means falla sistemáticamente y DBSCAN tiene éxito.

**Parámetros clave de DBSCAN:**
- `eps` (ε): radio de vecindad — ¿qué tan lejos puede estar un punto para ser vecino?
- `min_samples`: mínimo de puntos en el radio `eps` para que un punto sea *core point*


```python
# Dataset artificial: dos lunas entrelazadas
X_moon, y_moon = make_moons(n_samples=400, noise=0.05, random_state=SEED)

# K-Means (falla: asume clusters esféricos/convexos)
km_moon = KMeans(n_clusters=2, random_state=SEED, n_init=10)
labels_km_moon = km_moon.fit_predict(X_moon)

# DBSCAN (éxito: conecta regiones densas sin importar la forma)
db_moon = DBSCAN(eps=0.2, min_samples=5)
labels_db_moon = db_moon.fit_predict(X_moon)

ari_km_moon = metrics.adjusted_rand_score(y_moon, labels_km_moon)
ari_db_moon = metrics.adjusted_rand_score(y_moon, labels_db_moon)
n_noise     = (labels_db_moon == -1).sum()

print('🌌 Comparación sobre dataset "lunas entrelazadas" (n=400, ruido=0.05):')
print(f'  K-Means ARI:  {ari_km_moon:.3f}  ← falla en formas no esféricas')
print(f'  DBSCAN  ARI:  {ari_db_moon:.3f}  ← detecta la forma correctamente')
print(f'  Outliers detectados por DBSCAN: {n_noise} ({n_noise/len(y_moon)*100:.1f}% del total)')
print()
print('  ARI=1.0 → recuperó perfectamente la estructura real')

# Visualización comparativa
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Clusters de Forma Arbitraria: K-Means vs DBSCAN', fontsize=13)

# Estructura real
axes[0].scatter(X_moon[:, 0], X_moon[:, 1], c=y_moon,
                cmap='coolwarm', s=30, edgecolors='black', linewidths=0.3)
axes[0].set_title('Estructura Real (2 lunas)', fontsize=11)

# K-Means
axes[1].scatter(X_moon[:, 0], X_moon[:, 1], c=labels_km_moon,
                cmap='coolwarm', s=30, edgecolors='black', linewidths=0.3)
axes[1].set_title(f'K-Means (k=2)\nARI={ari_km_moon:.2f}  ❌ Falla', fontsize=11)

# DBSCAN
axes[2].scatter(X_moon[:, 0], X_moon[:, 1], c=labels_db_moon,
                cmap='coolwarm', s=30, edgecolors='black', linewidths=0.3)
mask_noise = labels_db_moon == -1
if mask_noise.any():
    axes[2].scatter(X_moon[mask_noise, 0], X_moon[mask_noise, 1],
                    c='gray', s=40, marker='x', label=f'Outliers ({n_noise})', zorder=5)
    axes[2].legend(fontsize=9)
axes[2].set_title(f'DBSCAN (eps=0.2, min=5)\nARI={ari_db_moon:.2f}  ✅ Éxito', fontsize=11)

for ax in axes:
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print('\n💡 K-Means parte el espacio con fronteras de Voronoi (lineales) → no puede seguir la forma.')
print('   DBSCAN conecta puntos en regiones densas → detecta la forma de las lunas perfectamente.')
```

## 🎓 Resumen y Conclusiones

Tanto la **clasificación** como el **clustering** buscan separar instancias en grupos, pero difieren fundamentalmente en su enfoque:

| | Clasificación | Clustering |
|---|---|---|
| **Tipo** | Supervisado | No supervisado |
| **Datos** | Etiquetados | Sin etiquetas |
| **Propósito** | Predecir clases conocidas | Descubrir grupos desconocidos |
| **Evaluación** | Accuracy, F1, ROC-AUC | Silhouette, ARI, Inercia |
| **Algoritmos** | LR, KNN, DT, RF, Naive Bayes | K-Means, Hierarchical, DBSCAN, OPTICS |

### 🔑 Regla de oro

> **¿Tengo etiquetas?** → Clasificación | **¿No tengo etiquetas?** → Clustering

### 📌 Hallazgos empíricos del notebook

- En Iris, K-Means (sin etiquetas) recuperó bien la estructura aunque con algunos errores en la frontera Versicolor/Virginica.
- DBSCAN demostró su superioridad en datos de forma no esférica (lunas) donde K-Means falla completamente.
- El clustering jerárquico ofrece una visualización intuitiva (dendrograma) y no requiere especificar $k$ de antemano.

### 🚀 ¿Qué sigue?

- 📐 **PCA / Reducción de dimensionalidad**: técnica no supervisada complementaria al clustering
- 🌐 **Redes Neuronales**: clasificadores poderosos para imágenes y texto
- 🔍 **Detección de anomalías**: Isolation Forest, One-Class SVM
- 📊 **Método del codo y Silhouette**: para determinar el $k$ óptimo en K-Means
