## 🌲 ¿Qué es un Árbol de Decisión?

Un **Árbol de Decisión** es un modelo que simula el proceso de toma de decisiones humano: a partir de una serie de preguntas sobre los datos, va dividiendo el espacio de posibilidades hasta llegar a una conclusión (una clase).

```text
    ¿Glucosa > 120?
     /         \
   Sí           No
   |             |
¿BMI > 30?   No Diabetes
  /     \
Sí       No
|        |
Diabetes No Diabetes
```

Cada **nodo interno** representa una pregunta sobre un atributo.  
Cada **rama** representa una respuesta posible.  
Cada **hoja** representa una predicción (clase final).

## 📦 Paso 1: Importar las Librerías

Antes de empezar, necesitamos importar las herramientas que vamos a usar:

- 🐼 **pandas**: para cargar y manipular datos
- 🌳 **DecisionTreeClassifier**: el algoritmo principal
- ✂️ **train_test_split**: para dividir los datos en entrenamiento y prueba
- 📊 **metrics**: para evaluar el rendimiento del modelo
- 🎨 **matplotlib / plot_tree**: para visualizar el árbol

````python
import pandas as pd
import matplotlib.pyplot as plt

# 🌳 Algoritmo de Árbol de Decisión
from sklearn.tree import DecisionTreeClassifier, plot_tree

# ✂️ División de datos
from sklearn.model_selection import train_test_split

# 📊 Métricas de evaluación
from sklearn import metrics

print('✅ Librerías importadas correctamente!')
````

## 📂 Paso 2: Cargar y Explorar los Datos

Cargamos el dataset CSV y le asignamos nombres descriptivos a las columnas.  
Luego hacemos una exploración inicial para entender con qué datos estamos trabajando.

> 💡 **Tip:** Siempre es importante explorar el dataset antes de entrenar un modelo. ¡Conocer tus datos es la mitad del trabajo!

````python
# Nombres de las columnas del dataset
col_names = ['pregnant', 'glucose', 'bp', 'skin',
             'insulin', 'bmi', 'pedigree', 'age', 'label']

# 📥 Cargar el dataset
pima = pd.read_csv('https://raw.githubusercontent.com/CristianPacifico/ia-ls-fcad-uner/main/data/ml/pima-indians-diabetes.csv', header=0, names=col_names)

print('📋 Primeras 5 filas del dataset:')
pima.head()
````

````python
# 🔍 Información general del dataset
print(f'📐 Dimensiones: {pima.shape[0]} filas x {pima.shape[1]} columnas\n')
print('📊 Distribución de la variable objetivo (label):')
print(pima['label'].value_counts())
print()
print('🔢 Estadísticas descriptivas:')
pima.describe()
````

````python
# 🔍 ¿Hay valores nulos?
print('❓ Valores nulos por columna:')
print(pima.isnull().sum())

# Visualizar la distribución de clases
fig, ax = plt.subplots(figsize=(5, 4))
counts = pima['label'].value_counts()
ax.bar(['No Diabetes (0)', 'Diabetes (1)'], counts.values,
       color=['steelblue', 'tomato'], edgecolor='black')
ax.set_title('Distribución de Clases', fontsize=14)
ax.set_ylabel('Cantidad de pacientes')
for i, v in enumerate(counts.values):
    ax.text(i, v + 5, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
````

## 🎯 Paso 3: Selección de Features (Variables)

En Machine Learning llamamos **features** (características) a las variables de entrada que el modelo usa para aprender.

- **X** → Matriz de features (variables independientes)
- **y** → Vector objetivo / etiquetas (variable dependiente)

Para este modelo usaremos las columnas más relevantes médicamente: `pregnant`, `insulin`, `bmi`, `age`, `glucose`, `bp` y `pedigree`.

> ⚠️ **Nota:** Excluir columnas irrelevantes o redundantes puede mejorar el rendimiento del modelo y reducir el sobreajuste (*overfitting*).

````python
# 🎯 Seleccionar las features (variables de entrada)
feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']

X = pima[feature_cols]   # Features (entrada)
y = pima['label']        # Target (salida: 0 o 1)

print('📥 Features (X) - primeras 5 filas:')
print(X.head())
print(f'\n📐 Forma de X: {X.shape}')

print('\n🎯 Target (y) - primeras 5 valores:')
print(y.head())
````

## ✂️ Paso 4: División en Entrenamiento y Prueba

Para evaluar correctamente un modelo de ML, necesitamos **separar los datos** en dos conjuntos:

| Conjunto | Proporción | Uso |
|----------|-----------|-----|
| **Entrenamiento** (train) | 70% | El modelo aprende de estos datos |
| **Prueba** (test) | 30% | Evaluamos qué tan bien generaliza el modelo |

> 🚨 **Regla de oro:** El modelo **nunca debe ver** los datos de prueba durante el entrenamiento.
> Si entrenamos y probamos con los mismos datos, no sabemos si el modelo realmente aprendió o solo memorizó.

El parámetro `random_state=6` garantiza que la división sea **reproducible** (siempre obtenemos la misma partición).

````python
# ✂️ Dividir en 70% entrenamiento y 30% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,      # 30% para prueba
    random_state=6      # Semilla para reproducibilidad
)

print('📊 Tamaños de los conjuntos:')
print(f'  🏋️  Entrenamiento (X_train): {X_train.shape[0]} muestras')
print(f'  🧪  Prueba       (X_test) : {X_test.shape[0]} muestras')
print(f'\n  Total: {X_train.shape[0] + X_test.shape[0]} muestras')
````

## 🌳 Paso 5: Crear y Entrenar el Árbol de Decisión

Ahora llegamos al corazón del notebook: **crear y entrenar el modelo**.

Con `scikit-learn`, esto se hace en tan solo dos líneas:
1. **Crear** la instancia del clasificador
2. **Entrenar** con `.fit(X_train, y_train)`

### 🔧 ¿Cómo aprende el árbol?

El algoritmo busca en cada paso la **mejor pregunta** para dividir los datos, usando un criterio como:
- **Gini Impurity**: mide qué tan mezcladas están las clases en un nodo
- **Entropy**: mide el desorden o incertidumbre de un nodo

Por defecto, `DecisionTreeClassifier` usa **Gini Impurity**.

````python
# 🌳 Crear el clasificador de Árbol de Decisión
model = DecisionTreeClassifier(random_state=42)

# 🏋️ Entrenar (ajustar) el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

print('✅ Modelo entrenado exitosamente!')
print(f'\n📐 Profundidad del árbol: {model.get_depth()}')
print(f'🍃 Número de hojas: {model.get_n_leaves()}')
````

## 🔮 Paso 6: Hacer Predicciones

Una vez entrenado el modelo, podemos usarlo para **predecir** la clase de nuevos datos (los del conjunto de prueba).

El método `.predict()` recorre el árbol para cada muestra y devuelve la clase predicha (0 o 1).

````python
# 🔮 Predecir las clases para el conjunto de prueba
y_pred = model.predict(X_test)

print('🔮 Primeras 10 predicciones:')
print('  Predicho: ', y_pred[:10])
print('  Real:     ', y_test.values[:10])

# Comparación visual rápida
aciertos = (y_pred[:10] == y_test.values[:10]).sum()
print(f'\n  ✅ Aciertos en las primeras 10: {aciertos}/10')
````

## 📊 Paso 7: Evaluación del Modelo — Métricas

Una vez que tenemos las predicciones, necesitamos **medir qué tan bueno es el modelo**.

### 🧩 La Matriz de Confusión

Todo parte de la **Matriz de Confusión**, que compara las clases reales vs las predichas:

|                |  Predicho: NO  |  Predicho: Si |
| -------------- | :------------: | :-----------: |
| Real: No Diab. |    TN ✅       |    FP ❌      |
| Real: Diabetes |    FN ❌       |    TP ✅      |


Siendo que:

| Sigla | Nombre | Significado |
|-------|--------|-------------|
| **TP** | True Positive | Predijo Diabetes, y ERA Diabetes ✅ |
| **TN** | True Negative | Predijo No Diabetes, y NO ERA Diabetes ✅ |
| **FP** | False Positive | Predijo Diabetes, pero NO ERA Diabetes ❌ |
| **FN** | False Negative | Predijo No Diabetes, pero ERA Diabetes ❌ |

---

### 📐 Las 4 métricas principales

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$

$$\text{Precision} = \frac{TP}{TP + FP}$$

$$\text{Recall} = \frac{TP}{TP + FN}$$

$$\text{F1} = 2 \cdot \frac{Precision \times Recall}{Precision + Recall}$$

> 🩺 **En medicina:** el **Recall** (sensibilidad) es especialmente importante porque nos interesa no perder ningún caso positivo de diabetes (FN peligroso).

````python
# 📊 Calcular todas las métricas
accuracy  = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred)
recall    = metrics.recall_score(y_test, y_pred)
f1        = metrics.f1_score(y_test, y_pred)

print('=' * 45)
print('       📊 MÉTRICAS DE EVALUACIÓN')
print('=' * 45)
print(f'  🎯 Accuracy   (exactitud) : {accuracy:.4f}  ({accuracy*100:.1f}%)')
print(f'  🔬 Precision  (precisión) : {precision:.4f}  ({precision*100:.1f}%)')
print(f'  📡 Recall     (sensibil.) : {recall:.4f}  ({recall*100:.1f}%)')
print(f'  ⚖️  F1 Score   (balance)   : {f1:.4f}  ({f1*100:.1f}%)')
print('=' * 45)

# También disponible como score directo del modelo
print(f'\n✅ Score directo del modelo (test): {model.score(X_test, y_test):.4f}')
````

````python
# 🧩 Visualizar la Matriz de Confusión
from sklearn.metrics import ConfusionMatrixDisplay

fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(
    y_test, y_pred,
    display_labels=['No Diabetes', 'Diabetes'],
    cmap='Blues',
    ax=ax
)
ax.set_title('Matriz de Confusión', fontsize=14, pad=15)
plt.tight_layout()
plt.show()
````

## 🌲 Paso 8: Visualizar el Árbol de Decisión

Una de las grandes ventajas de los Árboles de Decisión es que son **interpretables**: podemos visualizar exactamente qué preguntas hace el modelo para cada predicción.

> 💡 **Nota:** Este árbol crece sin límite de profundidad, por lo que puede ser grande. En la siguiente celda también veremos una versión **podada** (con `max_depth`) para mejor visualización.

````python
# 🌲 Visualizar el árbol completo (las primeras capas)
fig, ax = plt.subplots(figsize=(20, 8))

plot_tree(
    model,
    feature_names=feature_cols,
    class_names=['No Diabetes', 'Diabetes'],
    filled=True,
    rounded=True,
    max_depth=4,        # Mostramos solo los primeros 4 niveles para legibilidad
    fontsize=7,
    ax=ax
)

ax.set_title('Árbol de Decisión (primeros 4 niveles)', fontsize=16, pad=20)
plt.tight_layout()
plt.show()

print('💡 Los nodos azules indican predicción "No Diabetes"')
print('💡 Los nodos naranja/marrón indican predicción "Diabetes"')
print('💡 Cuanto más intenso el color, más pura es la clase en ese nodo')
````

## 🏆 Paso 9: Importancia de las Features

El árbol de decisión también nos dice **cuáles variables fueron más importantes** para tomar decisiones.

Esto se llama **Feature Importance** y se calcula en base a cuánto reduce la impureza (Gini) cada feature a lo largo del árbol.

> 🩺 ¿Qué variables del perfil médico predicen mejor la diabetes?

````python
# 🏆 Feature Importance
importances = pd.Series(model.feature_importances_, index=feature_cols)
importances_sorted = importances.sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(8, 5))
colors = plt.cm.RdYlGn(importances_sorted / importances_sorted.max())
importances_sorted.plot(kind='barh', ax=ax, color=colors, edgecolor='black')

ax.set_title('Importancia de las Features', fontsize=14)
ax.set_xlabel('Importancia relativa')

for i, v in enumerate(importances_sorted):
    ax.text(v + 0.002, i, f'{v:.3f}', va='center', fontsize=10)

plt.tight_layout()
plt.show()

print(f'\n🥇 Feature más importante: {importances.idxmax()} ({importances.max():.3f})')
print(f'🥉 Feature menos importante: {importances.idxmin()} ({importances.min():.3f})')
````

## 🔬 Bonus: ¿Qué pasa con el Overfitting?

Un Árbol de Decisión sin restricciones puede **memorizar** el dataset de entrenamiento: esto se llama **overfitting** (sobreajuste).

Veamos la diferencia entre la accuracy en entrenamiento vs prueba:

> 🚨 Si la accuracy en entrenamiento es mucho mayor que en prueba, el modelo está sobreajustado.

````python
# 🔬 Comparar accuracy en train vs test
acc_train = model.score(X_train, y_train)
acc_test  = model.score(X_test, y_test)

print('🔬 Diagnóstico de Overfitting:')
print(f'  🏋️  Accuracy en entrenamiento: {acc_train:.4f} ({acc_train*100:.1f}%)')
print(f'  🧪  Accuracy en prueba:        {acc_test:.4f} ({acc_test*100:.1f}%)')
print(f'\n  📉 Diferencia: {(acc_train - acc_test)*100:.1f} puntos porcentuales')

if acc_train - acc_test > 0.1:
    print('  ⚠️  ¡Posible OVERFITTING! Considera podar el árbol con max_depth.')
else:
    print('  ✅ El modelo generaliza bien.')
````

````python
# 🌲 Comparar modelos con diferente profundidad máxima
depths = range(1, 12)
train_scores = []
test_scores  = []

for d in depths:
    m = DecisionTreeClassifier(max_depth=d, random_state=42)
    m.fit(X_train, y_train)
    train_scores.append(m.score(X_train, y_train))
    test_scores.append(m.score(X_test, y_test))

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(depths, train_scores, 'o-', label='Entrenamiento', color='steelblue', linewidth=2)
ax.plot(depths, test_scores,  's-', label='Prueba',          color='tomato',    linewidth=2)
ax.set_xlabel('Profundidad máxima (max_depth)', fontsize=12)
ax.set_ylabel('Accuracy', fontsize=12)
ax.set_title('Accuracy vs Profundidad del Árbol\n(Bias-Variance Tradeoff)', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_xticks(list(depths))
plt.tight_layout()
plt.show()

best_depth = depths[test_scores.index(max(test_scores))]
print(f'\n🏆 Mejor profundidad (según test): max_depth={best_depth} → Accuracy={max(test_scores):.4f}')
````
