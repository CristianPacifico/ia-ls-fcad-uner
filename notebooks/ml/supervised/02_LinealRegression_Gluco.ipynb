## Regresión Lineal con el dataset de diabetes ('s6' - nivel de glucosa)

Este notebook muestra un ejemplo de regresión lineal utilizando el dataset de diabetes de Scikit-learn.

Se utiliza una sola variable numérica predictora (`s6`: nivel de glucosa en sangre) para predecir la progresión de la enfermedad. EL target es una medida cuantitativa de la progresión de la enfermedad un año después del inicio. 

## 1. Importar librerías necesarias

Importamos las librerías requeridas para el análisis, modelado y visualización de datos.

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

## 2. Cargar y explorar el dataset de diabetes

Cargamos el dataset de diabetes desde scikit-learn y exploramos las primeras filas para conocer las variables disponibles.

```python
# Cargar el dataset de diabetes
dtset = datasets.load_diabetes(return_X_y=False, as_frame=True)

# Mostrar las primeras filas del DataFrame
display(dtset['frame'].head())

# Mostrar los nombres de las columnas (características)
display("Características disponibles:", dtset['data'].columns.tolist())
```

## 3. Seleccionar la característica 's6'

Seleccionamos únicamente la columna 's6' (nivel de glucosa) como variable predictora y mostramos algunos ejemplos.

```python
# Seleccionar la característica 's6' (nivel de glucosa)
diabetes_X = dtset['data'][['s6']]
diabetes_y = dtset['target']

# Mostrar ejemplos de los primeros valores
print("Primeros valores de 's6':")
display(diabetes_X.head(20))
print("Primeros valores del target:")
display(diabetes_y.head(20))
```

## 4. Dividir los datos en conjuntos de entrenamiento y prueba

Dividimos los datos en conjuntos de entrenamiento y prueba usando train_test_split.

```python
X_train, X_test, y_train, y_test = train_test_split(
    diabetes_X, diabetes_y, test_size=0.1, random_state=1
)
print(f"Tamaño del conjunto de entrenamiento: {X_train.shape[0]}")
print(f"Tamaño del conjunto de prueba: {X_test.shape[0]}")
```

## 5. Entrenar el modelo de regresión lineal

Creamos el objeto LinearRegression y lo entrenamos con los datos de entrenamiento.

```python
# Crear el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Coeficientes hallados
print("Coeficiente:", model.coef_)
print("Intercepto:", model.intercept_)
type(model)
```

## 6. Realizar predicciones y evaluar el modelo

Predecimos los valores del conjunto de prueba y calculamos el score, los coeficientes, el error cuadrático medio y el coeficiente de determinación.

```python
# Realizar predicciones
y_pred = model.predict(X_test)

# Evaluar el modelo
print("Score del modelo:", model.score(X_test, y_test))
print("Promedio de error absoluto (MSE):", mean_absolute_error(y_test, y_pred))
print("Coeficiente de determinación (R²):", r2_score(y_test, y_pred))
```

## Analizamos los resultados de las métricas obtenidas

Si el coeficiente de determinación (R²) es mayor a 0.5, significa que el modelo explica más del 50% de la variabilidad del target.

Por el contrario, si es menor a 0.5, indica que el modelo no es capaz de explicar adecuadamente la variabilidad del target.

En este caso , la variable 's6' por sí sola NO es un buen predictor para la progresión de la enfermedad.  
El valor de R² es bajo, lo que indica que la mayor parte de la variabilidad del target no se explica solo con 's6'. 
Por ende, sería recomendable usar más variables para mejorar el modelo.

¿Podríamos probar con otra variable, o combinación de variables?

## 7. Visualizar los resultados

Graficamos los valores reales y predichos junto con la línea de regresión usando matplotlib.

```python
plt.figure(figsize=(8, 6))
# Línea de regresión
plt.plot(X_test, y_pred, color="blue", linewidth=1, label="Regresión lineal")
# Puntos reales
plt.scatter(X_test, y_test, color="black", label="Datos reales")
# Puntos predichos
plt.scatter(X_test, y_pred, color="red", label="Predicciones")
plt.xlabel('s6 (nivel de glucosa)')
plt.ylabel('Progresión de la enfermedad')
plt.title('Regresión lineal: s6 vs Progresión de la enfermedad')
plt.legend()
plt.show()
```
